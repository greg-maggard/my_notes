### Relational Plots 

```python
# scatterplot in matplotlib

plt.scatter(tips.total_bill, tips.tip)

#let's break it down by gender as we would in matplotlib (below we can see how we can do this more easily in Seaborn)
plt.scatter(tips[tips.sex == 'Female'].total_bill, tips[tips.sex == 'Female'].tip)
plt.scatter(tips[tips.sex == 'Male'].total_bill, tips[tips.sex == 'Male'].tip)
```

```python
# Explore relationship between total_bill and tip using relplot

sns.relplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# Update defaults (rc params in matplotlib)

sns.set_theme()  # https://seaborn.pydata.org/generated/seaborn.set_theme.html
```

```python
# same plot above, but with different sns defaults ()

#In Seaborn, rather than writing code for two separate plots for the differences in sex, we can simply pass the "hue" argument, and tell it to differentiate based on sex.

sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

### hue, size and style arguments

```python
# visualize if relationship is different for smoker vs non-smoker (use hue argument)

#you can also set different styles for different styles by passing the 'style' argument (you can also add size to differentiate some other dimention)

sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker', style = 'smoker', size = 'smoker')

#Ravinder's Advice: Adding a bunch of these arguments can get visually messy pretty quickly, so it may be best to stick to measuring one or two things at once. 
```

## Relplot with 'kind' argument

```python
# lineplot for total bill vs tips - may not be appropriate in this case. Note use of 'kind' argument


sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'line')
```

```python
# lineplot for discrete values. The lineplot will use mean value at each size to draw a line 
# shaded region is the 95% confidence interval. Can be turned off with 'ci = None'

sns.relplot(data = tips, x = 'size', y = 'tip', kind = 'line')
```

### Small multiple (https://en.wikipedia.org/wiki/Small_multiple)
- similar graphs or charts using the same scale and axes
- easy to compare


- Main idea: Pick a categorical features and create a chart for each category

```python
#  FacetGrid with 'col' 
#col_order lets you specify the order of the table via a list of the columns
# col_wrap allows you to designate how many columns you want ot display before it moves to another line

sns.relplot(data = tips, x = 'total_bill',
            y = 'tip',
            col = 'day',
            hue = 'time',
            col_order = ['Thur', 'Fri', 'Sat', 'Sun'],
            col_wrap = 2
           )
```

```python
# Facetgrid with lineplot

sns.relplot(data = tips, x = 'size', y = 'tip', kind = 'line', col = 'day')
```

#### Key takeaways?
- Tip amount generally increases with total_bill
- Waiter works in evenings on Sun and sat
- Waiter works during lunchtime on Thursday

#### Using Scatterplot or lineplots instead of relplot

```python
# scatterplot - returns an axes level object

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

## Distributions: displot

```python
# displot for total_bill (hist, kde, rug)
#kde is 'kerneld density estimation', which shows a smoothed-out representation of the distribution within a sample 
# the rug plot places a tick for ever data point, which will show the distribution of results, and can highlight the presence of points on the outlying edges of a dataset (think a tick for Jeff Bezos on the far right of an income distribution)
sns.displot(data = tips, x = 'total_bill', kde = True, rug = True)
```

```python
# histogram with hue and stacking and palette


sns.displot(data = tips, x = 'total_bill', hue= 'sex', multiple = 'stack', palette = ['r', 'b'])
```

```python
# use 'col' argument with 'sex' to create 'small multiples'

#note that 'alpha' represents the transparency of the bars

sns.displot(data = tips, x = 'total_bill', col = 'sex', color = 'blue', alpha = 1)
```

```python
#histplot will return an axes level object

sns.histplot(data = tips, x = 'total_bill')
```

#### Heatmap

```python
#crosstab of time vs smoker
ctab = pd.crosstab(tips.time, tips.smoker)
ctab
```

```python
#create a heatmap
#always be sure to include the 'cmap' argument (using '_r' at the end of the cmap value will reverse the spectrum)
#Keep in mind that generally, you want to map a higher number with a darker shade, and a lower numbers with a lighter shade.

sns.heatmap(ctab, cmap = 'Blues', annot = True, square = True, fmt = '0.0f')
```

Seaborn color palettes: https://seaborn.pydata.org/tutorial/color_palettes.html

- SEQUENTIAL: e.g. different shades of same color.  Appropriate when data range from relatively low or uninteresting values to relatively high or interesting values (or vice versa)
- DIVERGING - highlight both low and high values

```python
# Heatmap using mpg dataset

mpg = data('mpg')
mpg.head()
```

```python
# calculate correlation for numeric variables using panda's .corr() method

mpg.corr()
```

```python
labels = ['Displacement', 'Model Year', 'cylinders', 'City MPG', 'Highway MPG']
```

```python
# heatmap for correlation table above

sns.heatmap(mpg.corr(), annot = True,
           fmt = '0.2f', cmap = 'RdBu_r', 
           center = 0, vmin = -1, 
           linewidth = 1,
           linecolor = 'k',
           xticklabels = labels, 
           yticklabels = labels)
```
